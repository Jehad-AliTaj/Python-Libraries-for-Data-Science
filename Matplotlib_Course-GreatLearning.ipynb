# Line Plot

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
x= np.arange(1,11)
```

```python
x
```

```python
y= x*2
y
```

```python
plt.plot(x,y)
plt.show()

#as x increases Y is also increasing
```

```python
# Adding title and labels

plt.plot(x,y)
plt.title("Line Plot - Great Learning ")
plt.xlabel("X-label")
plt.ylabel("Y-label")
plt.show()
```

```python
# Changing line aesthetics

plt.plot(x,y, color='g', linestyle= ':', linewidth= 3)
plt.show()
```

```python
#Adding two lines in the same plot
y1= x*2
y2= x*3
plt.plot(x,y1, color= 'g', linestyle= ":", linewidth= " 3")
plt.plot(x,y1, color= 'r', linestyle= ":", linewidth= " 4")
plt.title("Line Plot - Great Learning ")
plt.xlabel("X-label")
plt.ylabel("Y-label")
plt.show()
```

```python
y1= x*2
y2= x*3
plt.plot(x,y1, color= 'g', linestyle= ":", linewidth= " 3")
plt.plot(x,y2, color= 'r', linestyle= ":", linewidth= " 4")
plt.title("Line Plot - Great Learning ")
plt.xlabel("X-label")
plt.ylabel("Y-label")
plt.grid(True)    #For showing grid as weel
plt.show()
```

```python
# Adding sub-plots

x= np.arange(1,11)
y1=x*2
y2=x*3
plt.subplot(1,2,1)
plt.plot(x,y1, color= 'g', linestyle= ":", linewidth= " 3")
plt.subplot(1,2,2)
plt.plot(x,y2, color= 'r', linestyle= ":", linewidth= " 4")
plt.show
```

# Bar Plot and Horizontal Bar Plot

## Bar Plot

```python
student = { "Tariq":80, "Dawood":70, "Nasir":60}
names= list(student.keys())
values = list(student.values())
plt.bar(names,values)
plt.show()

# First Parameter will have Catagorical values and the second will have Numerical 
#values
```

```python
# Adding Title and Labels

plt.bar(names,values)
plt.title("Bar Plot")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
# Horizontal Bar Plot

plt.barh(names,values)
plt.title("Bar Plot")
plt.xlabel("Marks")
plt.ylabel("Names")
# plt.grid(True)
plt.show()
```

```python
plt.barh(names,values, color= "red")
plt.title("Bar Plot")
plt.xlabel("Marks")
plt.ylabel("Names")
# plt.grid(True)
plt.show()
```

# Scatter Plot

```python
# Creating a basic Scatter Plot

x= [10,20,30,40,50,60,70,80,90]
a= [8,1,7,2,0,3,7,3,2]
plt.scatter(x,a)
plt.show()
```

```python
#Adding Marker
x= [10,20,30,40,50,60,70,80,90]
a= [8,1,7,2,0,3,7,3,2]
plt.scatter(x,a, marker= "*", c= "g", s=100)
plt.show()
```

```python
#Adding two markers in the same plot
x= [10,20,30,40,50,60,70,80,90]
a= [8,1,7,2,0,3,7,3,2]
b=[ 9,8,7,5,2,6,1,5,8]
# plt.scatter(x,a)
plt.scatter(x,a, marker= "*", c= "g", s=100)
plt.scatter(x,b, marker= ".", c= "r", s=200)
plt.show()
```

```python
# Adding Sub-plots
x= [10,20,30,40,50,60,70,80,90]
a= [8,1,7,2,0,3,7,3,2]
b=[ 9,8,7,5,2,6,1,5,8]
plt.subplot(1,2,1)
plt.scatter(x,a, marker= "*", c= "g", s=100)

plt.subplot(1,2,2)
plt.scatter(x,b, marker= ".", c= "r", s=200)
plt.show()
```

# Pie- Chart


```python
fruit=["Apple","Orange","Mango",  "Banana"] 
quantity= [80,70,60,50]
plt.pie(quantity,labels=fruit)
plt.show()
```

```python
# Changing Colors
fruit=["Apple","Orange","Mango",  "Banana"] 
quantity= [80,70,60,50]
plt.pie(quantity,labels=fruit, autopct='%0.1f%%', colors=['green','yellow','orange','pink'])
plt.show()
```

# DoughNut Chart

```python
fruit=["Apple","Orange","Mango",  "Banana"] 
quantity= [80,70,60,50]
plt.pie(quantity,labels=fruit, radius=2)
plt.pie([1],colors=['w'] ,radius=1)
plt.show()
```
